```python
# import variables..
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#loading the dataset...
data=pd.read_csv('heart.csv')
data
```

```python
data.columns.values
```

```python
#summary of the dataset..
data.info()
```

```python
# checking for null values..
data.isnull().sum()
```

```python
#plotting histogram of all numeric values..
data.hist(bins=100, grid=True, figsize=(15,15))
```

```python
#Generating descriptive stastics..
data.describe()
```

```python
Questions=["1.How many people have heart disease and how many people doesnt have.?",
           "2.People of which sex has most heart disease.?",
           "3.People of which sex has which type of chest  pain most.?",
           "4.People with which chest pain to have heart disease.?"]
Questions
```

```python
# Plotting Bar Chart
data.target.value_counts().plot(kind='bar', color=["orchid","salmon"])
plt.title("Heart Disease Values")
plt.xlabel("1 =Heart Disease, 0 =No Heart Disease")
plt.ylabel("Amount")
plt.show()
```

```python
#1.How many people have heart disease and how many people doesnt have.?
#getting the values
data.target.value_counts()
```

```python
#Plotting a Pie Chart..
data.target.value_counts().plot(kind='pie', figsize=(6,6))
plt.legend(['Disease',"No Disease"])
plt.show()
```

```python
#  Number of Male and Female in the dataset..
data.sex.value_counts()
```

```python
# plotting a bar chart..
data.sex.value_counts().plot(kind='bar')
plt.title('Male Female Ratio')
plt.legend(['Male','Female'])
```

```python
# plotting a pie chart..
data.sex.value_counts().plot(kind='pie',figsize=(8,6))
plt.title('Male Female Ratio')
plt.legend(['Male','Female'])
```

```python
#"2.People of which sex has most heart disease.?"
pd.crosstab(data.target,data.sex)
```

```python
sns.countplot(x='target', data=data,hue='sex')
plt.title('Heart Disease Frequency for Sex')
plt.xlabel('0=No Heart Disease, 1=Heart Disease')
```

```python
#3.People of which sex has which type of chest  pain most.?
data.cp.value_counts()
```

```python
#plotting a bar chart...
data.cp.value_counts().plot(kind='bar',color=['salmon','lightskyblue','springgreen','khaki'])
plt.title('Chest pain type VS Count')
```

```python
sns.countplot(x='cp',data=data, hue='target')
```

```python
# Normal Distribution Curve..
sns.displot(x='age',data=data, bins=30,kde=True)
```

```python
# Distribution plot for "Maximum Heart Rate"
sns.displot(x='thalach', data=data, bins=30, kde=True, color='chocolate')
```
